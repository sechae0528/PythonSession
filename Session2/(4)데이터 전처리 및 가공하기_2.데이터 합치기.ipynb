# 1. 추가 속성값을 추가하는 경우 (join)

#### 데이터 조인 : pd. merge(데이터프레임명1, 데이터프레임명2, left_on=‘데이터프레임1 조인 키’
####                        right_on=‘데이터프레임2 조인 키’, how = “left”)  
##### (how 파라미터 기본값은 inner)

```python
import numpy as np

# 데이터 조인
pd.merge(selectExample, regionMasterData,left_on='regionid', right_on = 'regionid')\
[['regionid','regionname','product','yearweek','qty']]
```

#### group by 평균을 -> 이름이라는 데이터프레임에 담고, kopo_channel_seasonality_new 라는 데에 데이터프레임을 담는다.

### group by( regionid, product 까지 그룹바이하고), 평균값으로 넣어서 다시 만들기

```python
# 평균 구하기
meanValue = selectExample.\
groupby(['regionid','product'], as_index=False)[["qty"]].\
mean() #평균구하는 함수

meanValue.head()
```

```python
meanValue = meanValue.reset_index()
meanValue.head()
```

```python
# 데이터 조인
selectExample1 = pd.merge(selectExample, meanValue,on = ['regionid','product'],\
                          suffixes=('_a', '_b'), how="left")
selectExample1.head()
```

```python
selectExample1["ratio"] = selectExample1["qty_a"]/selectExample1["qty_b"]
selectExample1.head()
```

```python
#컬럼명 전체 변경
selectExample1.columns = \
["REGIONID","PRODUCT","YEARWEEK", \
 "QTY", "QTY_AVG", "RATIO"]

selectExample1.head()
```

# 2. 동일한 컬럼 데이터를 합치는 경우 (union)

###### axis = 1 옆으로 붙일때, 0이면 아래로 붙일때

#### 데이터 조인 : pd. concat([데이터프레임명1, 데이터프레임명2])

```python
import numpy as np

# A01 데이터 정제
a01data = selectExample[(selectExample['regionid']=='A01')]
print(len(a01data))
print(a01data.count())

# A02 데이터 정제
a02data = selectExample[(selectExample['regionid']=='A02')]
a02data['qty'] = np.where(a02data['qty'] < 500,500,a02data['qty'])
print(len(a02data))
print(a02data.count())

# 데이터 결과 값 MERGE
mergedata = pd.concat([a01data, a02data], axis=0)
print(len(mergedata))
print(mergedata.count())
```

# 3. 피벗 활용하기

### 1. PIVOT (행을 열로)

#### pivot : 데이터프레임명.pivot_table( index = [‘고정key컬럼1’,’고정key컬럼2’], 
####  columns=‘pivot대상컬럼명’, values=‘피벗산대상컬럼’)

```python
#  피벗 예시
pivotResult = selectExample.\
pivot_table(index = ['regionid','product'], columns='yearweek', values='qty')
pivotResult.head()
```

```python
finalResult = pivotResult.reset_index()
finalResult.head()
```

```python
finalResult.columns
```

### 2. UNPIVOT (열을 행으로)

### unpivot : 데이터프레임명.stack().reset_index(

```python
unpivotResult=pivotResult.stack().reset_index()
unpivotResult.columns=\
['regionid','product','yearweek','qty']
unpivotResult.head()
```

```python
import requests, bs4, pandas as pd
#웹페이지 html 소스 가져오기
resp = requests.get('http://finance.daum.net/quote/marketvalue.daum?nil_stock=refresh')
resp.encoding='utf-8'
html = resp.text

bs = bs4.BeautifulSoup(html, 'html.parser')
bs
```

```python
contents=bs.select('ul#myListTop1 > li', limit=7)
print(contents)
dfcontent=[]
alldfcontents=[]
for content in contents:
    tds=content.findAll("a")
    for td in tds:
        if(td.text=="None"):
            continue;
        dfcontent.append(td.text)
        print(dfcontent)
    alldfcontents.append(dfcontent)
    #print(alldfcontents)
    dfcontent=[]
df=pd.DataFrame(data=alldfcontents)
df
```

```python
contents=bs.select('ul#myListTop1 > li', limit=7)
#print(contents)
dfcontent1=[]
alldfcontents1=[]
for content in contents:
    tds=content.findAll("span",{"class": "cost num"})
#    print(tds)
    for td in tds:
        if(td.text=="None"):
            continue;
        dfcontent1.append(td.text)    
       # print(dfcontent1)
    alldfcontents1.append(dfcontent1)
    #print(alldfcontents1)
    dfcontent1=[]
df1=pd.DataFrame(data=alldfcontents1)
df1
```

```python
finalDf = pd.concat([df,df1],axis=1)
finalDf.columns = ['stock_name','stock_price']
finalDf
```

```python
finalDf.to_csv("../dataset/kopo_stock_st_cse.csv", encoding='utf-8-sig',index=False)
```

```python
import requests, bs4, pandas as pd
#웹페이지 html 소스 가져오기
resp = requests.get('http://finance.daum.net/quote/marketvalue.daum?nil_stock=refresh')
resp.encoding='utf-8'
html = resp.text

bs = bs4.BeautifulSoup(html, 'html.parser')
stock_list=bs.select('ul#myListTop1 > li', limit=7)
#stock_list = bs.select("ul,{"class":"myListTop1}).findAll("li")
stock_list

count = 0
stock_nameList =[]
stock_priceList = []
stock_num = 6

for stock_one in stock_list:
    if count > stock_num:
        break
    stock_name = stock_one.find("a").text
    stock_price = float(stock_one.find("span",{"class":"cost num"}).text.replace(",",""))
    #stock_price = stock_one.find("span",{"class":"cost num"}).text
    stock_nameList.append(stock_name)
    stock_priceList.append(stock_price)
    count += 1
list1 = pd.DataFrame(stock_nameList, dtype=str)
list2 = pd.DataFrame(stock_priceList, dtype=float)
final = pd.concat([list1,list2], axis=1)
final.columns=['STOCK_NAME', 'STOCK_PRICE']
final
```

```python
from sqlalchemy import create_engine 
# DB 커넥션 열기
engine = create_engine('oracle+cx_oracle://kopo:kopo@192.168.110.112:1521/orcl') 


# 데이터 저장
tablename='KOPO_STOCK_ST_채성은'
final.to_sql(tablename, engine, if_exists='replace', index=False)
```

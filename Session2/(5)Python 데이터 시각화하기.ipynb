## 2차원 차트 Subplot

```python
#필요 라이브러리 정의
import numpy as np
import matplotlib.pyplot as plt

#데이터 정의 첫번째 컬럼은 정규분포, 두번째 컬럼은 *100
y=np.random.standard_normal((20,2))
y[:,1] = y[:,1]*100

plt.figure(figsize=(7,5))
plt.subplot(121) #1행2열에 첫번째다
plt.plot(y[:, 0], 'b',lw=1.5, label = '1st')
#plt.plot(y[:, 0], 'o')
plt.grid(True)
plt.legend(loc=0)
plt.ylabel('value')
plt.title('A Simple Plot')
plt.subplot(122) #1행2열에 두번째다
plt.plot(y[:, 1], 'g', lw=1.5, label='2nd')
#plt.plot(y[:, 1], 'go')
plt.grid(True)
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
```

```python
#필요 라이브러리 정의
import numpy as np
import matplotlib.pyplot as plt

#데이터 정의 첫번째 컬럼은 정규분포, 두번째 컬럼은 *100
y=np.random.standard_normal((20,2))
y[:,1] = y[:,1]*100

plt.figure(figsize=(7,5))
plt.subplot(211) #2행1열에 첫번째다
plt.plot(y[:, 0], 'b',lw=1.5, label = '1st')
plt.plot(y[:, 0], 'o')
plt.grid(True)
plt.legend(loc=0)
plt.ylabel('value')
plt.title('A Simple Plot')
plt.subplot(212) #2행1열에 두번째다
plt.plot(y[:, 1], 'g', lw=1.5, label='2nd')
plt.plot(y[:, 1], 'go') #g는 green 색 o는 점을 뜻함
plt.grid(True)
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
```

### 2차원 차트 그리기

```python
#필요 라이브러리 정의
import numpy as np
import matplotlib.pyplot as plt

#데이터 정의 첫번째 컬럼은 정규분포, 두번째 컬럼은 *100
y=np.random.standard_normal((20,2))
y[:,1] = y[:,1]*100

#첫번째 컬럼자료를 첫번째(왼쪽) 축에 대응하여 그림
fig, ax1 = plt.subplots()
plt.plot(y[:, 0], 'b', lw=1.5, label = '1st')
plt.plot(y[:, 0], 'bo')
plt.grid(True)
plt.legend(loc=4)
plt.xlabel('index')
plt.ylabel('value 1st')
plt.title('A Simple Plot')
#두번째 컬럼자료를 두번쨰(오른쪽) 축에 대응하여 그림
ax2 = ax1.twinx()
plt.plot(y[:, 1], 'g', lw=1.5, label = '2nd')
plt.plot(y[:, 1], 'go')
plt.legend(loc=0)
plt.ylabel('value 2nd')
```

## Distplot 차트 구현
#### 특정 컬럼 값의 분포도를 확인하기 위한 차트

```python
#pip install seasborn
import seaborn as sns
%matplotlib inline

tips = sns.load_dataset('tips')
tips.head()
```

```python
#분포도 구현
import seaborn as sns
%matplotlib inline

sns.distplot(tips['total_bill'])
```

## Jointplot 차트 구현
#### 데이터셋에 대하여 특정 2가지 값에 대하여 그래프형태로 표현

```python
#두가지 값에 대한 분포
#금액대비 실제 tip 분포 및 각 카테고리별 분포도를 함께 볼수 있음
sns.jointplot(x= 'total_bill', y = 'tip', data = tips)
```

## pairplot 차트 구현
#### 전체 데이터의 모든관계를 한눈에 보기위한 차트

```python
sns.pairplot(tips)

sns.pairplot(tips, hue = 'sex') #sex 그룹바이한 것 
```

## barplot 차트 구현
#### 카테고리 데이터에 대하여 구분하는 차트

```python
sns.barplot(x='sex', y='total_bill', data = tips)
```

## countplot 차트 구현
#### 해당개수를 세어주는 countplot

```python
sns.countplot(x= 'sex', data = tips)
```

## boxplot 차트 구현
#### 정해진 카테고리 기준에 어떤 범위의 분포를 보여주는지 차트화함

```python
sns.boxplot(x='day', y='total_bill',data=tips)
```

## corr
#### 데이터의 상관관계를 확인

```python
tips.corr()
```

## Heatmap
#### 자료에 대한 상관관계를 heat map으로 표현

```python
tc = tips.corr()
sns.heatmap(tc, annot = True)
```

```python
flights = sns.load_dataset('flights')
fp = flights.pivot_table(index='month', columns='year', values = 'passengers')

fp.head()
```

## Implot
#### 두 데이터의 상관관계를 linear한 형태로 뽑아내는 차트

```python
sns.lmplot(x='total_bill', y = 'tip', data = tips )
sns.lmplot(x='total_bill', y = 'tip', data = tips, hue = 'sex', markers = ['o','v'] )
```

###  컬럼별 결측값 개수확인하기

```python
sampleData.isnull().sum()
```

```python
sampleData['qty'].isnull().sum()
```

```python
sampleData.notnull().sum()
```

```python
sampleData['NaN_cnt'] = sampleData.isnull().sum(1)
```

### 결측값 채우기

#### 1. 특정 값으로 채우기

```python
sampleData.fillna(0)
```

#### 2. 특정 순서로 채우기

```python
sampleData
```

```python
sampleData.fillna(method="bfill")
```

```python
sampleData.fillna(method="ffill")
```

#### 3. 평균 값으로 채우기

```python
sampleData
```

```python
sampleData.fillna(sampleData.mean())
```

```python
sampleData
```

```python
sampleData.where(pd.notnull(sampleData), sampleData.mean(), axis='columns')
```

```python
sampleData.fillna(sampleData.mean()['qty':'target'])
```

#### 4. 다른 컬럼의 값으로 채우기

```python
sampleData['new_qty'] = np.where(pd.notnull(sampleData['qty']) == True, sampleData['qty'], sampleData['target'])
sampleData['new_target'] = np.where(pd.notnull(sampleData['target']) == True, sampleData['target'], sampleData['qty'])
sampleData
```

```python
sampleData.fillna(method="ffill",inplace=True)
```

```python
sampleData.fillna(method="bfill",inplace=True)
```

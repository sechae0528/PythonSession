# 4. 이상 데이터 정제하기

#### 데이터프레임명[‘컬럼명‘] = np.where( 조건식, yes값, no값)

```python
# numpy 라이브러리 
import numpy as np

#컬럼 추가 생성 및 조건에 따른 컬럼명 변경
customerData["PRODUCTAGE_NEW"] = np.\
where(customerData["PRODUCTAGE"] < 1, 1, customerData["PRODUCTAGE"])

customerData.head()
```

### 예제: productage가 1보다작으면 1 2보다작으면2 3보다작으면3 이외는 5라고 새로운 컬럼 product_age_new를 생성하세요

```python
import numpy as np

#step = [1,2,3]
step = list(range(1,4))
maxValue = 5

customerData["PRODUCT_AGE_NEW"] = np.where(customerData["PRODUCTAGE"] < step[0], step[0], \
                                            np.where(customerData["PRODUCTAGE"]< step[1],step[1],\
                                            np.where(customerData["PRODUCTAGE"]< step[2],step[2],\
                                            maxValue)))
customerData.head(10)
```

```python
#컬럼 추가 생성 및 조건에 따른 컬럼명 변경
def refining(df):
    if df['PRODUCTAGE'] < 1:
        return 1
    elif df['PRODUCTAGE'] < 2:
        return 2
    elif df['PRODUCTAGE'] < 3:
        return 3
    else:
        return 5

customerData["PRODUCTAGE_NEW2"] =customerData.apply(refining, axis=1)
customerData.head(10)
```

# 6. 데이터 요약분석하기(group by)

## rolling

```python
customerData["TEST"] = customerData["PRODUCTAGE"].rolling(window=5, center=True).mean()
#center가 True이면 이동평균, 그 값 중심으로 양옆 2개씩을 합한 5개의 평균을 구함
#center가 False이면 앞에 4개포함 자기자신까지의 평균을 나타냄.
customerData
```

```python
import math
suborder = math.floor(5/2)
suborder
```

```python
list2 = [1,2]
```

```python
for i in range(0, suborder):
    customerData.loc[i,"TEST"] = list2[i]
customerData.head()    
```

```python
for i in range(0,suborder):
    customerData["TEST"][i] = list2[i]
customerData.head()
```

```python
first_list = []
for i in range(0,suborder):
    first_list.append(customerData["PRODUCTAGE"][0:i+suborder].mean())
first_list
```

```python
last_list = []
for i in range(0,suborder):
    last_list.append(customerData["PRODUCTAGE"][i-suborder-1:].mean())
last_list
```

```python
for i in range(0,suborder):
    customerData["TEST"][i] = first_list[i]
    
for i in range(0,suborder):
    customerData["TEST"][len(customerData)-(1+i)] = last_list[i]

customerData
```

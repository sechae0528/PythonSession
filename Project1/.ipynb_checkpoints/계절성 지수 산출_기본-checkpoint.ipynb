### 3. 이동평균 구하기(Rolling 연산)

```python
import math

# 이동평균 구간
order = 13

# 앞뒤로 비어있게 될 구간 크기
suborder = math.floor(order/2)
suborder
```

```python
# rolling 연산 설명: 이동평균 구간의 중간값을 기준으로 앞뒤로 비어있게 될 구간에 해당하는 범위에 수치형 컬럼인 QTY 평균값을 넣는다.
# 예를 들어 5의 중간값인 3을 기준으로 위쪽 1,2  + 중간값 3 + 아래쪽 4,5까지 총 5 구간의 평균을 구하는 것  

sortedData["MA"] = sortedData["QTY"].rolling(window = order, center =True).mean() #13이 들어가는지 확인
```

### 4. 이동평균 빈셀 채우기

```python
maxLength = len(sortedData)
maxLength
```

- 앞쪽 빈칸 채우기

```python
prelist=[] # 앞단
for i in range(0,suborder):
    prelist.append(sortedData["QTY"][0:i+suborder].mean())
    sortedData.loc[i,"MA"]= prelist[i]
print(prelist)
sortedData.head(13)
```

- 뒤쪽 빈칸 채우기

```python
postList = [] #뒤단
maxLength = len(sortedData)
for i in range(0,suborder):
    postList.append(sortedData["QTY"][-(i+suborder+1):maxLength].mean())    
    sortedData.loc[maxLength-i-1,"MA"]= postList[i]
print(postList)
sortedData.tail(15)
```

- 계절성 지수 산출

```python
sortedData["RATIO"] = sortedData["QTY"]/sortedData["MA"]
sortedData.head()
```

### 5. 그룹핑 + 롤링함수

```python
# 하지만 이동평균이 잘 구해진것 같지만 롤링이 제대로 안되었다.
# 이동평균을 그룹핑 안하고 돌리니까, ST0001 -> ST0002로 넘어갈때, 갑자기 값이 달라진다.

# GROUR BY 해서 롤링하는 방법을 알아야 한다.
```

```python
def sub_function(data):
    data["MA"] = data["QTY"].rolling(window = order, center =True).mean()
    return data
```

```python
# apply 적용하면 groupby 된 data 가 자동으로 함수 전달인자로 들어간다.
groupResult = sortedData.groupby(["REGIONID", "PRODUCT"]).apply(sub_function)
groupResult.head()
```

```python
#인덱스 초기화
grouping=groupResult.reset_index(drop=True)
grouping.head()
```

```python
#DataFrame의 isnull 함수를 이용해 결측치를 확인한다
grouping.isnull()
```

```python
#Data Frame의 fillna함수를 사용해 전체 프레임의 결측치를 상수 0으로 채워 보정
grouping.fillna(0,inplace="True")
grouping
```
